```python
import json
import os

FILE_JSON = "inventory.json"

# ---------------- Fungsi Load & Save JSON ----------------
def load_inventory():
    if os.path.exists(FILE_JSON):
        with open(FILE_JSON, "r") as file:
            return json.load(file)
    return []

def save_inventory(data):
    with open(FILE_JSON, "w") as file:
        json.dump(data, file, indent=4)

inventory = load_inventory()

# ---------------- Fitur Inventory ----------------
def tampilkan_barang():
    if not inventory:
        print("üì≠ Inventory kosong.\n")
        return
    print("\nDaftar Barang:")
    print("No | Nama Barang | Stok | Harga")
    print("-" * 40)
    for i, barang in enumerate(inventory, 1):
        print(f"{i}. {barang['nama']} | {barang['stok']} | Rp{barang['harga']}")
    print()

def tambah_barang():
    nama = input("Masukkan nama barang: ")
    stok = int(input("Masukkan jumlah stok: "))
    harga = float(input("Masukkan harga satuan: "))
    inventory.append({"nama": nama, "stok": stok, "harga": harga})
    save_inventory(inventory)
    print(f"‚úÖ Barang '{nama}' berhasil ditambahkan!\n")

def edit_barang():
    tampilkan_barang()
    if not inventory:
        return
    no = int(input("Masukkan nomor barang yang ingin di-edit: "))
    if 1 <= no <= len(inventory):
        barang = inventory[no - 1]
        print(f"Edit barang: {barang['nama']}")
        barang['nama'] = input("Nama baru (kosongkan jika tidak diubah): ") or barang['nama']
        stok = input("Stok baru (kosongkan jika tidak diubah): ")
        harga = input("Harga baru (kosongkan jika tidak diubah): ")
        if stok:
            barang['stok'] = int(stok)
        if harga:
            barang['harga'] = float(harga)
        save_inventory(inventory)
        print("‚úÖ Data barang berhasil diperbarui!\n")
    else:
        print("‚ö†Ô∏è Nomor barang tidak valid.\n")

def hapus_barang():
    tampilkan_barang()
    if not inventory:
        return
    no = int(input("Masukkan nomor barang yang ingin dihapus: "))
    if 1 <= no <= len(inventory):
        barang = inventory.pop(no - 1)
        save_inventory(inventory)
        print(f"üóëÔ∏è Barang '{barang['nama']}' berhasil dihapus.\n")
    else:
        print("‚ö†Ô∏è Nomor barang tidak valid.\n")

def cari_barang():
    keyword = input("Masukkan nama barang yang dicari: ").lower()
    hasil = [b for b in inventory if keyword in b['nama'].lower()]
    if hasil:
        print("\nüîç Hasil Pencarian:")
        for b in hasil:
            print(f"{b['nama']} | Stok: {b['stok']} | Rp{b['harga']}")
    else:
        print("‚ùå Barang tidak ditemukan.\n")

def laporan_ringkas():
    total_stok = sum(b["stok"] for b in inventory)
    total_nilai = sum(b["stok"] * b["harga"] for b in inventory)
    print(f"\nüìä Laporan Ringkas:")
    print(f"- Total jumlah stok    : {total_stok}")
    print(f"- Total nilai inventory: Rp{total_nilai:,.2f}\n")

# ---------------- Menu Utama ----------------
def menu():
    while True:
        print("""
=== APLIKASI INVENTORY (JSON) ===
1. Tambah Barang
2. Lihat Daftar Barang
3. Edit Barang
4. Hapus Barang
5. Cari Barang
6. Laporan Ringkas
0. Keluar
""")
        pilihan = input("Pilih menu: ")
        if pilihan == "1":
            tambah_barang()
        elif pilihan == "2":
            tampilkan_barang()
        elif pilihan == "3":
            edit_barang()
        elif pilihan == "4":
            hapus_barang()
        elif pilihan == "5":
            cari_barang()
        elif pilihan == "6":
            laporan_ringkas()
        elif pilihan == "0":
            print("üëã Terima kasih telah menggunakan aplikasi ini.")
            break
        else:
            print("‚ö†Ô∏è Pilihan tidak valid.\n")

if __name__ == "__main__":
    menu()
```
